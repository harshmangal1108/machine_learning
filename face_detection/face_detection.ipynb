```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
image=plt.imread("teeam.jpg")
```

```python
plt.imshow(image)
```

```python
image.shape
```

```python
detector=cv2.CascadeClassifier("haarcascade_frontalface_alt (1).xml")
```

```python
detector
```

```python
faces=detector.detectMultiScale(image)
```

```python
faces
```

```python
count=0
for face in faces:
    x,y, w, h = face
    count+=1
    image = cv2.rectangle(image, (x,y), (x+w, y+h), (40,220,150), 3)
```

```python
plt.imshow(image)
print("Number of faces:",count)
```
